```python
x = data.drop(['diagnosis'],axis=1)
y = data.diagnosis
x.columns
```

```python
##creating training and test data
from sklearn.model_selection import train_test_split
x_train_raw,x_test_raw,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10)
print(x_train_raw.shape,x_test_raw.shape,y_train.shape,y_test.shape)
```

```python
## Scaling the data as the magnitude of variables are varying lot
from sklearn.preprocessing import StandardScaler
sc=StandardScaler() 
x_train=sc.fit_transform(x_train_raw)
x_test=sc.transform(x_test_raw) 
y_train.value_counts()
#B : 282   M: 173
```

```python
## model creation
import sklearn
from sklearn.neural_network import MLPClassifier
model = MLPClassifier( hidden_layer_sizes=(50,4),
                       learning_rate_init=0.001,
                       max_iter=150,
                       random_state=2) ## model object creation max_iter=Stopping parameter

model.fit(x_train,y_train) ## training the data

y_predict_proba = model.predict_proba(x_test)

y_pred = model.predict(x_test)
y_pred1 = model.predict(x_train)

sklearn.set_config(print_changed_only=False)
model
```

```python
#Evaluation
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
acc = accuracy_score(y_test,y_pred)
pr = precision_score(y_test,y_pred,pos_label= 1)
re = recall_score(y_test,y_pred,pos_label= 1)
f1 = f1_score(y_test,y_pred,pos_label= 1)
cm = confusion_matrix(y_test,y_pred)
print('Accuracy Score: ',acc,'\nPrecision:', pr,'\nRecall Score:', re,'\nF1 Score', f1, '\nConfusion Matrix: \n', cm)
print(classification_report(y_test,y_pred))
```
